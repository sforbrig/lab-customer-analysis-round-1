```python
import pandas as pd
import numpy as np

from IPython.display import display
```

```python
file1 = pd.read_csv('file1.csv')

display(file1.head())
display(file1.tail())

file1.shape
```

```python
file2 = pd.read_csv('file2.csv')

display(file2.head())
display(file2.tail())

file2.shape
```

```python
file3 = pd.read_csv('file3.csv')

display(file3.head())
display(file3.tail())

file3.shape
```

```python
display(file1.columns)
display(file2.columns)
display(file3.columns)
```

```python
file1 = file1.rename(columns={'ST':'State',
                            'GENDER':'Gender'})

cols = []
for column in file1.columns:
    cols.append(column.lower())
file1.columns = cols

file1.head()
```

```python
file2 = file2.rename(columns={'ST':'State',
                            'GENDER':'Gender'})

cols = []
for column in file2.columns:
    cols.append(column.lower())
file2.columns = cols

file2.head()
```

```python
cols = []
for column in file3.columns:
    cols.append(column.lower())
file3.columns = cols

file3.head()
```

```python
display(file1.columns)
```

```python
file2new = file2[['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]

file3new = file3[['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]

display(file1.columns)
display(file2new.columns)
display(file3new.columns)
```

```python
fileX = pd.concat([file1, file2new, file3new], axis=0)
fileX.shape
fileX
```

```python
fileX.dtypes
```

```python
fileXnew = fileX.drop(['education','number of open complaints'], axis =1)
fileXnew
```

```python
fileXnew['customer lifetime value'] = fileXnew['customer lifetime value'].replace('%', '', regex=True)
fileXnew.head()
```

```python
fileXnew['customer lifetime value'] = pd.to_numeric(fileXnew['customer lifetime value'], errors='coerce')
fileXnew
```

```python
fileXnew['customer lifetime value'] = fileXnew['customer lifetime value'] /100
fileXnew
```

```python
print(fileXnew.shape)
fileY = fileXnew.copy()
fileY = fileY.drop_duplicates()
print(fileY.shape)
fileY
```
